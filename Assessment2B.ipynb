###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
data_exceeded = df[df["CodeKata Score"]>15000]
data_exceeded.to_csv("Exceeded_expectations.csv")
data_exceeded
```

###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
data_expected = df[(df["CodeKata Score"]>10000)&(df["CodeKata Score"]<15000)]
data_expected.to_csv("Reached_expectations.csv")
data_expected
```

###### if  7000<codekata score<10000   (Needs_Improvement.csv)


```python
data_toimprove = df[(df["CodeKata Score"]>7000)&(df["CodeKata Score"]<10000)]
data_toimprove.to_csv("Needs_Improvement.csv")
data_toimprove
```

###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
data_unsatisfied = df[df["CodeKata Score"]<7000]
data_unsatisfied.to_csv("Unsatisfactory.csv")
data_unsatisfied
```
